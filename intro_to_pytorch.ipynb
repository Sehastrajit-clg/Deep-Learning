**Basic operations using Pytorch**

```python
import torch as t
import numpy as np
```

```python
data=[[1,2],[3,4]]
x_data=t.tensor(data)
```

**From NumPy array**

```python
np_array=np.array(data)
x_np=t.from_numpy(np_array)
x_np
```

**Random Tensor**

```python
x_rand=t.rand_like(x_data,dtype=t.float)
x_rand
```

**Attributes of a tensor-shape,data type and the device on which they are stored**

```python
tensor=t.rand(3,4)
print("Shape of tensor",tensor.shape)
print("Datatype of the tensor ",tensor.dtype)
print("Device tensor is stored on",tensor.device)
```

**From CPU to GPU**

```python
if t.cuda.is_available():
  tensor=tensor.to("cuda")
print("Now tensor is stored on:",t.device)
```

**Concatenation**

```python
concat=t.cat([x_data,x_data],dim=1)
concat
```

**Arithematic operations**

**Matrix multiplication**

```python
matrix_A=t.tensor([[1,2],[3,4]], dtype=t.float32)
matrix_B=t.tensor([[1,2],[3,4]], dtype=t.float32)

result=t.matmul(matrix_A,matrix_B)
result
```

**Eigen value Eigen vector**

```python
eigenvalues,eigenvectors=t.linalg.eig(matrix_A)
print("Eigen values:\n",eigenvalues)
print("Eigen vectors:\n",eigenvectors)
```
